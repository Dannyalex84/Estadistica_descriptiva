```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values)


covarience_matrix = np.cov(scaled.T)
covarience_matrix
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'])
sns.jointplot(x = scaled[:, 2], y = scaled[:, 3])
plt.show()
```

#Descomposicion en vectores y valores propios:
https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html

```python
eigen_values, eigen_vectors = np.linalg.eig(covarience_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
variance_explained = []
for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)

print(variance_explained)
```

PCA con scikit : https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
reduced_scaled = pca.transform(scaled)
```

```python
reduced_scaled
```
